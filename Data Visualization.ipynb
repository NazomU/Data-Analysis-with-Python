```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker = '^', markersize = 10, linestyle = '--')

z = [1,5,10]
plt.plot(x,z, label = 'Baths', color = 'red', marker = '.', markersize = 10, linestyle = '--')


plt.title('Matplotlib Viz', fontdict = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```

```python
housing = pd.read_csv(r"C:\Users\nazom\OneDrive\Documents\Python Scripts\Pandas Folder\real_estate.csv", encoding = "ISO-8859-1")
housing
```

```python
housing.corr()
```

```python
housing_filtered = housing.drop('zip_code', axis = 1)
```

```python
correlation = housing_filtered.corr()
```

```python
sns.heatmap(correlation, annot = True, cmap = 'coolwarm')
```

```python
pearson_correlation = housing_filtered.corr(method = 'pearson')
pearson_correlation
```

```python
spearman_correlation = housing_filtered.corr(method = 'spearman')
spearman_correlation
```

```python
kendall_correlation = housing_filtered.corr(method = 'kendall')
kendall_correlation
```

```python
sns.heatmap(pearson_correlation, annot = True, cmap = 'coolwarm')
plt.title('Pearson Correlation')
plt.show()
```

```python
sns.heatmap(spearman_correlation, annot = True, cmap = 'coolwarm')
plt.title('Spearman Correlation')
plt.show()
```

```python
sns.heatmap(kendall_correlation, annot = True, cmap = 'coolwarm')
plt.title('Kendall Correlation')
plt.show()
```
